```python
import os
import numpy as np
import json
import glob
```

```python
filelocation = os.path.join('azobenzene', 'rhf_dihedral_scan','*.json')
filenames = glob.glob(filelocation)
print(filenames)
```

```python

# Opening JSON file
f = open('data.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
 
# Iterating through the json
# list
for i in data['emp_details']:
    print(i)
 
# Closing file
f.close()
```

```python
dihedral_vals = []
energy_vals = []
for file in filenames:
    f = open(file)
    data = json.load(f)
    ev = data["cavity_free_rhf_energy"]
    dv = data["constrained_dihedral"]
    dihedral_vals.append(dv)
    energy_vals.append(ev)
    #print(data["molecule"]["constrained_dihedral"])
    #print(data["constrained_dihedral"])
    #print(data["cavity_free_rhf_energy"])
```

```python
from matplotlib import pyplot as plt
plt.plot(dihedral_vals, energy_vals, 'ro')
#plt.xlim(100, 180)
```
