```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import griddata

# Conversion factor from Hartrees to kcal/mol
HARTREE_TO_KCALMOL = 627.5094740631

def load_and_analyze_dataset(df, dataset_name, n_grid_points=24):
    """
    Perform interpolation and accuracy analysis on a molecular energy dataset.
    
    Parameters:
    -----------
    df : pandas DataFrame
        DataFrame with columns 'theta', 'phi', 'value'
    dataset_name : str
        Name of the dataset for labeling plots
    n_grid_points : int
        Number of grid points for interpolation in each dimension
    
    Returns:
    --------
    dict : Dictionary containing interpolation results and statistics
    """
    
    print(f"\n{'='*60}")
    print(f"Processing: {dataset_name}")
    print(f"{'='*60}")
    
    # Extract values
    theta_vals = df['theta'].values
    phi_vals = df['phi'].values
    values = df['value'].values
    
    print(f"Data points: {len(values)}")
    print(f"Theta range: [{theta_vals.min():.4f}, {theta_vals.max():.4f}]")
    print(f"Phi range: [{phi_vals.min():.4f}, {phi_vals.max():.4f}]")
    print(f"Value range: [{values.min():.6e}, {values.max():.6e}]")
    
    # Create pivot table
    pivot_df = df.pivot_table(index='theta', columns='phi', values='value')
    pivot_df.index = pivot_df.index.round(4)
    pivot_df.columns = pivot_df.columns.round(4)
    
    # Create fine grid for interpolation
    # phi (x-axis): 0-360 degrees, theta (y-axis): 0-180 degrees
    phi_grid = np.linspace(phi_vals.min(), phi_vals.max(), n_grid_points)
    theta_grid = np.linspace(theta_vals.min(), theta_vals.max(), n_grid_points)
    PH, TH = np.meshgrid(phi_grid, theta_grid)
    
    # Perform interpolation on grid
    grid_values = griddata((phi_vals, theta_vals), values, (PH, TH), method='linear')
    
    # Test accuracy: interpolate at original data points
    interpolated_at_original = griddata((phi_vals, theta_vals), values,
                                       (phi_vals, theta_vals), method='linear')
    
    # Calculate errors
    absolute_errors = np.abs(values - interpolated_at_original)
    relative_errors = np.abs((values - interpolated_at_original) / values) * 100
    
    # Remove NaN values (edge points that couldn't be interpolated)
    valid_mask = ~np.isnan(interpolated_at_original)
    absolute_errors_clean = absolute_errors[valid_mask]
    relative_errors_clean = relative_errors[valid_mask]
    values_clean = values[valid_mask]
    
    # Print statistics
    print(f"\nInterpolation Accuracy Assessment:")
    print(f"{'-'*60}")
    print(f"Successfully interpolated: {np.sum(valid_mask)}/{len(values)}")
    print(f"\nAbsolute Error Statistics:")
    print(f"  Mean: {np.mean(absolute_errors_clean):.6e}")
    print(f"  Std:  {np.std(absolute_errors_clean):.6e}")
    print(f"  Max:  {np.max(absolute_errors_clean):.6e}")
    print(f"  Min:  {np.min(absolute_errors_clean):.6e}")
    print(f"\nRelative Error Statistics (%):")
    print(f"  Mean: {np.mean(relative_errors_clean):.4f}%")
    print(f"  Std:  {np.std(relative_errors_clean):.4f}%")
    print(f"  Max:  {np.max(relative_errors_clean):.4f}%")
    
    # Find worst interpolation points
    worst_indices = np.argsort(absolute_errors_clean)[-5:]
    valid_indices = np.where(valid_mask)[0]
    
    print(f"\n5 Worst Interpolation Points:")
    print(f"{'-'*60}")
    for i, idx in enumerate(worst_indices):
        original_idx = valid_indices[idx]
        print(f"{i+1}. Theta={theta_vals[original_idx]:.4f}, Phi={phi_vals[original_idx]:.4f}")
        print(f"   Original: {values[original_idx]:.6f}, Interpolated: {interpolated_at_original[original_idx]:.6f}")
        print(f"   Error: {absolute_errors_clean[idx]:.6e}, Relative: {relative_errors_clean[idx]:.4f}%")
    
    # Create visualization
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))
    fig.suptitle(f'{dataset_name} - Interpolation Analysis', fontsize=16, fontweight='bold')
    
    # 1. Original data scatter plot
    scatter = axes[0, 0].scatter(df['phi'], df['theta'], c=df['value'], 
                                cmap='viridis', s=50, edgecolors='black', linewidth=0.5)
    axes[0, 0].set_xlabel('Phi (degrees)')
    axes[0, 0].set_ylabel('Theta (degrees)')
    axes[0, 0].set_title('Original Data Points')
    plt.colorbar(scatter, ax=axes[0, 0], label='Energy')
    
    # 2. Interpolated surface
    contour = axes[0, 1].contourf(PH, TH, grid_values, levels=20, cmap='viridis')
    axes[0, 1].scatter(phi_vals, theta_vals, c='red', s=10, alpha=0.5, label='Original points')
    axes[0, 1].set_xlabel('Phi (degrees)')
    axes[0, 1].set_ylabel('Theta (degrees)')
    axes[0, 1].set_title('Interpolated Surface')
    axes[0, 1].legend()
    plt.colorbar(contour, ax=axes[0, 1], label='Energy')
    
    # 3. Absolute error distribution
    axes[1, 0].hist(absolute_errors_clean, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
    axes[1, 0].set_xlabel('Absolute Error')
    axes[1, 0].set_ylabel('Frequency')
    axes[1, 0].set_title('Distribution of Absolute Errors')
    axes[1, 0].axvline(np.mean(absolute_errors_clean), color='red', linestyle='--', 
                       linewidth=2, label=f'Mean: {np.mean(absolute_errors_clean):.2e}')
    axes[1, 0].legend()
    
    # 4. Error vs original value
    axes[1, 1].scatter(values_clean, absolute_errors_clean, alpha=0.5, s=20, color='steelblue')
    axes[1, 1].set_xlabel('Original Energy Value')
    axes[1, 1].set_ylabel('Absolute Error')
    axes[1, 1].set_title('Error vs Original Value')
    axes[1, 1].set_yscale('log')
    axes[1, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return {
        'dataset_name': dataset_name,
        'pivot_df': pivot_df,
        'phi_grid': phi_grid,
        'theta_grid': theta_grid,
        'PH': PH,
        'TH': TH,
        'grid_values': grid_values,
        'mean_abs_error': np.mean(absolute_errors_clean),
        'mean_rel_error': np.mean(relative_errors_clean),
        'max_abs_error': np.max(absolute_errors_clean),
        'n_valid': np.sum(valid_mask),
        'n_total': len(values)
    }


def compute_and_analyze_difference(df1, df2, name1, name2, results1, results2):
    """
    Compute energy differences between two datasets and test interpolation accuracy.
    
    Parameters:
    -----------
    df1, df2 : pandas DataFrames
        DataFrames with 'theta', 'phi', 'value' columns
    name1, name2 : str
        Names of the datasets
    results1, results2 : dict
        Results dictionaries from load_and_analyze_dataset
    
    Returns:
    --------
    dict : Dictionary containing difference analysis results
    """
    
    diff_name = f"{name1} - {name2}"
    print(f"\n{'='*60}")
    print(f"Computing: {diff_name}")
    print(f"{'='*60}")
    
    # Merge datasets on theta and phi to ensure matching points
    merged = pd.merge(df1, df2, on=['theta', 'phi'], suffixes=('_1', '_2'))
    print(f"Matched data points: {len(merged)}")
    
    # Compute raw differences in kcal/mol
    merged['diff_raw'] = (merged['value_1'] - merged['value_2']) * HARTREE_TO_KCALMOL
    
    print(f"Raw difference range: [{merged['diff_raw'].min():.4f}, {merged['diff_raw'].max():.4f}] kcal/mol")
    print(f"Raw difference mean: {merged['diff_raw'].mean():.4f} kcal/mol")
    
    # Extract values
    theta_vals = merged['theta'].values
    phi_vals = merged['phi'].values
    diff_raw = merged['diff_raw'].values
    
    # Interpolate the difference directly
    n_grid = 24
    phi_grid = np.linspace(phi_vals.min(), phi_vals.max(), n_grid)
    theta_grid = np.linspace(theta_vals.min(), theta_vals.max(), n_grid)
    PH, TH = np.meshgrid(phi_grid, theta_grid)
    
    diff_interpolated_grid = griddata((phi_vals, theta_vals), diff_raw,
                                     (PH, TH), method='linear')
    
    # Compute difference of individually interpolated values at grid points
    diff_from_interp = (results1['grid_values'] - results2['grid_values']) * HARTREE_TO_KCALMOL
    
    # Test accuracy at original points
    # Method 1: Interpolate differences directly
    diff_interp_at_original = griddata((phi_vals, theta_vals), diff_raw,
                                      (phi_vals, theta_vals), method='linear')
    
    # Method 2: Compute difference from individually interpolated values
    value1_at_original = griddata((df1['phi'].values, df1['theta'].values), 
                                 df1['value'].values,
                                 (phi_vals, theta_vals), method='linear')
    value2_at_original = griddata((df2['phi'].values, df2['theta'].values),
                                 df2['value'].values,
                                 (phi_vals, theta_vals), method='linear')
    diff_from_indiv_interp = (value1_at_original - value2_at_original) * HARTREE_TO_KCALMOL
    
    # Calculate errors
    valid_mask1 = ~np.isnan(diff_interp_at_original)
    abs_err1 = np.abs(diff_raw - diff_interp_at_original)
    abs_err1_clean = abs_err1[valid_mask1]
    
    valid_mask2 = ~np.isnan(diff_from_indiv_interp)
    abs_err2 = np.abs(diff_raw - diff_from_indiv_interp)
    abs_err2_clean = abs_err2[valid_mask2]
    
    valid_mask3 = valid_mask1 & valid_mask2
    abs_err3 = np.abs(diff_interp_at_original - diff_from_indiv_interp)
    abs_err3_clean = abs_err3[valid_mask3]
    
    print(f"\nInterpolation Method Comparison:")
    print(f"{'-'*60}")
    print(f"Method 1: Direct interpolation of differences")
    print(f"  Mean abs error: {np.mean(abs_err1_clean):.6f} kcal/mol")
    print(f"  Max abs error:  {np.max(abs_err1_clean):.6f} kcal/mol")
    print(f"  Valid points:   {np.sum(valid_mask1)}/{len(diff_raw)}")
    
    print(f"\nMethod 2: Difference of individual interpolations")
    print(f"  Mean abs error: {np.mean(abs_err2_clean):.6f} kcal/mol")
    print(f"  Max abs error:  {np.max(abs_err2_clean):.6f} kcal/mol")
    print(f"  Valid points:   {np.sum(valid_mask2)}/{len(diff_raw)}")
    
    print(f"\nDifference between Method 1 and Method 2:")
    print(f"  Mean abs difference: {np.mean(abs_err3_clean):.6f} kcal/mol")
    print(f"  Max abs difference:  {np.max(abs_err3_clean):.6f} kcal/mol")
    
    # Visualization
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle(f'{diff_name} (kcal/mol) - Interpolation Analysis', 
                 fontsize=16, fontweight='bold')
    
    # Row 1: Interpolated surfaces
    cont1 = axes[0, 0].contourf(PH, TH, diff_interpolated_grid, levels=20, cmap='RdBu_r')
    axes[0, 0].scatter(phi_vals, theta_vals, c='black', s=5, alpha=0.3)
    axes[0, 0].set_xlabel('Phi (degrees)')
    axes[0, 0].set_ylabel('Theta (degrees)')
    axes[0, 0].set_title('Direct Interpolation of Differences')
    plt.colorbar(cont1, ax=axes[0, 0], label='kcal/mol')
    
    cont2 = axes[0, 1].contourf(PH, TH, diff_from_interp, levels=20, cmap='RdBu_r')
    axes[0, 1].set_xlabel('Phi (degrees)')
    axes[0, 1].set_ylabel('Theta (degrees)')
    axes[0, 1].set_title('Difference of Interpolated Surfaces')
    plt.colorbar(cont2, ax=axes[0, 1], label='kcal/mol')
    
    diff_methods = diff_interpolated_grid - diff_from_interp
    cont3 = axes[0, 2].contourf(PH, TH, diff_methods, levels=20, cmap='seismic')
    axes[0, 2].set_xlabel('Phi (degrees)')
    axes[0, 2].set_ylabel('Theta (degrees)')
    axes[0, 2].set_title('Difference Between Methods')
    plt.colorbar(cont3, ax=axes[0, 2], label='kcal/mol')
    
    # Row 2: Error analysis
    scatter = axes[1, 0].scatter(merged['phi'], merged['theta'], c=merged['diff_raw'],
                                cmap='RdBu_r', s=50, edgecolors='black', linewidth=0.5)
    axes[1, 0].set_xlabel('Phi (degrees)')
    axes[1, 0].set_ylabel('Theta (degrees)')
    axes[1, 0].set_title('Raw Differences at Data Points')
    plt.colorbar(scatter, ax=axes[1, 0], label='kcal/mol')
    
    axes[1, 1].hist(abs_err1_clean, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
    axes[1, 1].set_xlabel('Absolute Error (kcal/mol)')
    axes[1, 1].set_ylabel('Frequency')
    axes[1, 1].set_title('Error Distribution: Direct Interpolation')
    axes[1, 1].axvline(np.mean(abs_err1_clean), color='red', linestyle='--',
                      linewidth=2, label=f'Mean: {np.mean(abs_err1_clean):.4f}')
    axes[1, 1].legend()
    
    axes[1, 2].hist(abs_err2_clean, bins=30, edgecolor='black', alpha=0.7, color='coral')
    axes[1, 2].set_xlabel('Absolute Error (kcal/mol)')
    axes[1, 2].set_ylabel('Frequency')
    axes[1, 2].set_title('Error Distribution: Difference of Interpolations')
    axes[1, 2].axvline(np.mean(abs_err2_clean), color='red', linestyle='--',
                      linewidth=2, label=f'Mean: {np.mean(abs_err2_clean):.4f}')
    axes[1, 2].legend()
    
    plt.tight_layout()
    plt.show()
    
    return {
        'name': diff_name,
        'merged': merged,
        'PH': PH,
        'TH': TH,
        'diff_interpolated_grid': diff_interpolated_grid,
        'diff_from_interp': diff_from_interp,
        'mean_err_direct': np.mean(abs_err1_clean),
        'mean_err_diff_of_interp': np.mean(abs_err2_clean),
        'mean_diff_between_methods': np.mean(abs_err3_clean)
    }


# ===== LOAD ALL DATASETS =====

print("="*60)
print("LOADING DATASETS")
print("="*60)

# META dataset
print("\nLoading META dataset...")
meta_df = pd.read_csv('META/meta_ccsd_energies.txt', sep='\s+', header=None, 
                      names=['filename', 'value'])
pattern = r"theta_(.*?)_phi_(.*?)\.json"
meta_df[['theta', 'phi']] = meta_df['filename'].str.extract(pattern).astype(float)
meta_df = meta_df[['theta', 'phi', 'value']]
print(f"Loaded {len(meta_df)} rows.")

# PARA dataset
print("\nLoading PARA dataset...")
para_df = pd.read_csv('PARA/para_ccsd_energies.txt',
                      delim_whitespace=True,
                      header=None,
                      names=['theta', 'phi', 'value'])
print(f"Loaded {len(para_df)} rows.")

# ORTHO dataset
print("\nLoading ORTHO dataset...")
ortho_df = pd.read_csv('ORTHO/ortho_cc_files/ortho_ccsd_energies.txt',
                       delim_whitespace=True,
                       header=None,
                       names=['theta', 'phi', 'value'])
print(f"Loaded {len(ortho_df)} rows.")


# ===== ANALYZE ALL DATASETS =====

meta_results = load_and_analyze_dataset(meta_df, "META CCSD", n_grid_points=24)
para_results = load_and_analyze_dataset(para_df, "PARA CCSD", n_grid_points=24)
ortho_results = load_and_analyze_dataset(ortho_df, "ORTHO CCSD", n_grid_points=24)


# ===== COMPARISON SUMMARY =====

print(f"\n{'='*60}")
print("DATASET COMPARISON SUMMARY")
print(f"{'='*60}")

results = [meta_results, para_results, ortho_results]

print(f"\n{'Dataset':<15} {'Points':<10} {'Valid':<10} {'Mean Abs Err':<15} {'Mean Rel Err %':<15}")
print(f"{'-'*70}")
for r in results:
    print(f"{r['dataset_name']:<15} {r['n_total']:<10} {r['n_valid']:<10} "
          f"{r['mean_abs_error']:<15.6e} {r['mean_rel_error']:<15.4f}")

# Create comparison plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Interpolation Quality Comparison', fontsize=16, fontweight='bold')

for i, r in enumerate(results):
    axes[i].contourf(r['PH'], r['TH'], r['grid_values'], levels=20, cmap='viridis')
    axes[i].set_xlabel('Phi (degrees)')
    axes[i].set_ylabel('Theta (degrees)')
    axes[i].set_title(f"{r['dataset_name']}\nMean Abs Error: {r['mean_abs_error']:.2e}")

plt.tight_layout()
plt.show()


# ===== ENERGY DIFFERENCE ANALYSIS =====

print(f"\n{'='*60}")
print("ENERGY DIFFERENCE ANALYSIS")
print(f"{'='*60}")

# Compute differences
ortho_minus_meta = compute_and_analyze_difference(
    ortho_df, meta_df, "ORTHO", "META", ortho_results, meta_results
)

para_minus_meta = compute_and_analyze_difference(
    para_df, meta_df, "PARA", "META", para_results, meta_results
)


# ===== FINAL COMPARISON =====

print(f"\n{'='*60}")
print("DIFFERENCE INTERPOLATION SUMMARY")
print(f"{'='*60}")

diff_results = [ortho_minus_meta, para_minus_meta]

print(f"\n{'Difference':<20} {'Method 1 Err':<15} {'Method 2 Err':<15} {'Methods Diff':<15}")
print(f"{'-'*70}")
for r in diff_results:
    print(f"{r['name']:<20} {r['mean_err_direct']:<15.6f} "
          f"{r['mean_err_diff_of_interp']:<15.6f} {r['mean_diff_between_methods']:<15.6f}")

print("\nUnits: kcal/mol")
print("Method 1: Direct interpolation of differences")
print("Method 2: Difference of individually interpolated surfaces")

# Side-by-side comparison plot
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('Energy Differences (kcal/mol) - Comparison', fontsize=16, fontweight='bold')

# ORTHO - META
cont1 = axes[0, 0].contourf(ortho_minus_meta['PH'], ortho_minus_meta['TH'],
                           ortho_minus_meta['diff_interpolated_grid'], 
                           levels=20, cmap='RdBu_r')
axes[0, 0].set_xlabel('Phi (degrees)')
axes[0, 0].set_ylabel('Theta (degrees)')
axes[0, 0].set_title('ORTHO - META (Direct Interpolation)')
plt.colorbar(cont1, ax=axes[0, 0], label='kcal/mol')

cont2 = axes[0, 1].contourf(ortho_minus_meta['PH'], ortho_minus_meta['TH'],
                           ortho_minus_meta['diff_from_interp'],
                           levels=20, cmap='RdBu_r')
axes[0, 1].set_xlabel('Phi (degrees)')
axes[0, 1].set_ylabel('Theta (degrees)')
axes[0, 1].set_title('ORTHO - META (Diff of Interpolations)')
plt.colorbar(cont2, ax=axes[0, 1], label='kcal/mol')

# PARA - META
cont3 = axes[1, 0].contourf(para_minus_meta['PH'], para_minus_meta['TH'],
                           para_minus_meta['diff_interpolated_grid'],
                           levels=20, cmap='RdBu_r')
axes[1, 0].set_xlabel('Phi (degrees)')
axes[1, 0].set_ylabel('Theta (degrees)')
axes[1, 0].set_title('PARA - META (Direct Interpolation)')
plt.colorbar(cont3, ax=axes[1, 0], label='kcal/mol')

cont4 = axes[1, 1].contourf(para_minus_meta['PH'], para_minus_meta['TH'],
                           para_minus_meta['diff_from_interp'],
                           levels=20, cmap='RdBu_r')
axes[1, 1].set_xlabel('Phi (degrees)')
axes[1, 1].set_ylabel('Theta (degrees)')
axes[1, 1].set_title('PARA - META (Diff of Interpolations)')
plt.colorbar(cont4, ax=axes[1, 1], label='kcal/mol')

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("ALL ANALYSIS COMPLETE!")
print("="*60)


# ===== CREATE UNIFIED DATAFRAME =====

print(f"\n{'='*60}")
print("CREATING UNIFIED DATAFRAME")
print(f"{'='*60}")

# Round theta and phi values to handle small floating point differences
meta_df['theta_rounded'] = meta_df['theta'].round(6)
meta_df['phi_rounded'] = meta_df['phi'].round(6)
para_df['theta_rounded'] = para_df['theta'].round(6)
para_df['phi_rounded'] = para_df['phi'].round(6)
ortho_df['theta_rounded'] = ortho_df['theta'].round(6)
ortho_df['phi_rounded'] = ortho_df['phi'].round(6)

# Merge all three datasets on rounded theta and phi
print("\nMerging datasets...")
unified_df = meta_df[['theta_rounded', 'phi_rounded', 'value']].copy()
unified_df.rename(columns={'value': 'meta_energy'}, inplace=True)

# Merge PARA
unified_df = unified_df.merge(
    para_df[['theta_rounded', 'phi_rounded', 'value']],
    on=['theta_rounded', 'phi_rounded'],
    how='outer',
    suffixes=('', '_para')
)
unified_df.rename(columns={'value': 'para_energy'}, inplace=True)

# Merge ORTHO
unified_df = unified_df.merge(
    ortho_df[['theta_rounded', 'phi_rounded', 'value']],
    on=['theta_rounded', 'phi_rounded'],
    how='outer',
    suffixes=('', '_ortho')
)
unified_df.rename(columns={'value': 'ortho_energy'}, inplace=True)

# Rename rounded columns back to theta and phi
unified_df.rename(columns={'theta_rounded': 'theta', 'phi_rounded': 'phi'}, inplace=True)

# Calculate energy differences in kcal/mol
unified_df['ortho_minus_meta_kcalmol'] = (unified_df['ortho_energy'] - unified_df['meta_energy']) * HARTREE_TO_KCALMOL
unified_df['para_minus_meta_kcalmol'] = (unified_df['para_energy'] - unified_df['meta_energy']) * HARTREE_TO_KCALMOL

# Sort by theta and phi for readability
unified_df.sort_values(['theta', 'phi'], inplace=True)
unified_df.reset_index(drop=True, inplace=True)

# Print summary
print(f"\nUnified dataframe shape: {unified_df.shape}")
print(f"Total unique (theta, phi) pairs: {len(unified_df)}")
print(f"\nMissing values per column:")
print(unified_df.isnull().sum())

print(f"\nFirst few rows:")
print(unified_df.head(10))

print(f"\nDataframe statistics:")
print(unified_df.describe())

# Save to file
output_filename = 'unified_isomer_energies.csv'
unified_df.to_csv(output_filename, index=False, float_format='%.10f')
print(f"\n Unified dataframe saved to: {output_filename}")

# Also save as Excel for easier viewing (optional)
try:
    output_excel = 'unified_isomer_energies.xlsx'
    unified_df.to_excel(output_excel, index=False)
    print(f" Unified dataframe also saved to: {output_excel}")
except:
    print("  (Excel export not available - pandas excel dependencies not installed)")

# Check for any mismatches
print(f"\n{'='*60}")
print("DATA QUALITY CHECK")
print(f"{'='*60}")

# Check if any rows are missing data from specific isomers
meta_missing = unified_df['meta_energy'].isnull().sum()
para_missing = unified_df['para_energy'].isnull().sum()
ortho_missing = unified_df['ortho_energy'].isnull().sum()

print(f"\nData points missing by isomer:")
print(f"  META:  {meta_missing} missing")
print(f"  PARA:  {para_missing} missing")
print(f"  ORTHO: {ortho_missing} missing")

if meta_missing > 0 or para_missing > 0 or ortho_missing > 0:
    print("\n Warning: Some isomers have missing data points.")
    print("This could indicate different sampling grids were used.")
    
    # Show which theta/phi combinations are missing from which isomers
    if meta_missing > 0:
        print(f"\nTheta/Phi pairs missing from META:")
        print(unified_df[unified_df['meta_energy'].isnull()][['theta', 'phi']])
    if para_missing > 0:
        print(f"\nTheta/Phi pairs missing from PARA:")
        print(unified_df[unified_df['para_energy'].isnull()][['theta', 'phi']])
    if ortho_missing > 0:
        print(f"\nTheta/Phi pairs missing from ORTHO:")
        print(unified_df[unified_df['ortho_energy'].isnull()][['theta', 'phi']])
else:
    print("\n All isomers have complete data at all (theta, phi) points!")

print("\n" + "="*60)
print("UNIFIED DATAFRAME CREATION COMPLETE!")
print("="*60)
```
