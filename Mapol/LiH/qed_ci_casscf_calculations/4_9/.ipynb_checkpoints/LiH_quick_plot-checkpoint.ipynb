```python
import json
import numpy as np
from matplotlib import pyplot as plt
import os
os.pwd
```

```python


# Load the molecule data from the JSON file
with open("LiH_qedcas_4e_9o_10ph_lz_0.05_om_0.12086.json", 'r') as json_file:
    LiH_data = json.load(json_file)
```

```python
r = np.linspace(1.0, 3.0, 21)
E_casci = np.array(LiH_data["return_results"]["casci_energies"])
E_casscf = np.array(LiH_data["return_results"]["casscf_energies"])

print(np.shape(E_casci))
print(np.shape(E_casscf))
```

```python
plt.plot(r, E_casci[:,2])
plt.plot(r, E_casscf[:,2])
plt.plot(r, E_casci[:,1])
plt.plot(r, E_casscf[:,1])
```
