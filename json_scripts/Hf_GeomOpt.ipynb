```python
import numpy as np
from matplotlib import pyplot as plt
import psi4
```

```python
mol_str = """
H
F 1 0.9158
symmetry c1
"""

options_dict = {
    "basis": "6-311++G**",
    "scf_type": "pk",
    "e_convergence": 1e-10,
    "d_convergence": 1e-10,
    "frozen_docc" : [2],
    "active" : [16]
}

mol = psi4.geometry(mol_str)
psi4.set_options(options_dict)
scf_e, wfn = psi4.energy('SCF', return_wfn=True)
fci_energy, wfn = psi4.optimize('fci',ref_wfn=wfn, return_wfn=True)
```
